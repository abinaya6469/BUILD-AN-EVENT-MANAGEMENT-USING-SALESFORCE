# **LINEAR REGRESSION ANALYSIS OF CLOSING PRICES FOR ^IXIC**

```python

# Extract the relevant columns for linear regression Convert datetime index to
# numerical representation (e.g., days since the first date)
X = (sp500.index - sp500.index.min()).values.reshape(-1, 1)
y = sp500['Close'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the closing prices
y_pred = model.predict(X)

# Plot the actual and predicted closing prices
plt.figure(figsize=(12, 6))
plt.plot(sp500.index, y, label='Actual Close')
plt.plot(sp500.index, y_pred, label='Predicted Close')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Linear Regression Analysis of Closing Prices for ^IXIC')
plt.legend()
plt.grid(True)
plt.show()
```

**PLOT RESIDUAL HISTOGRAM AND DISTRIBUTION

```python
# Calculate residuals
residuals = y - y_pred

# Plot residual histogram
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Residual Histogram')
plt.grid(True)
plt.show()

# Plot residual distribution (Q-Q plot)
import statsmodels.api as sm
sm.qqplot(residuals, line='s')
plt.title('Residual Distribution (Q-Q Plot)')
plt.grid(True)
plt.show()
```

**PLOT NORMAL DISTRIBUTION FOR ^IXIC**

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Calculate mean and standard deviation of closing prices
mu = sp500['Close'].mean()
sigma = sp500['Close'].std()

# Generate x-values for the normal distribution curve
x = np.linspace(sp500['Close'].min(), sp500['Close'].max(), 100)

# Calculate corresponding y-values (probability density)
y = norm.pdf(x, mu, sigma)

# Plot the normal distribution curve
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.xlabel('Closing Price')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Closing Prices for ^IXIC')
plt.grid(True)
plt.show()
```

**Statistical summary of test data for ^IXIC**

```python

import yfinance as yf
# Fetch historical data for ^IXIC (NASDAQ Composite)
data = yf.download("^IXIC", start="2020-01-01", end="2023-01-01")

# Extract closing prices
close_prices = data['Close']

# Split the data into training and testing sets (80% train, 20% test)
train_data, test_data = train_test_split(close_prices, test_size=0.2, shuffle=False)

# Calculate statistical summary for test data
test_summary = test_data.describe()

print(test_summary)
```

```python
# prompt: WRITE A PROGRAM TO Calculate and print values of MAE, MSE, RMSE FOR ^IXIC

# Calculate daily returns for both training and testing data
train_returns = train_data.pct_change().dropna()
test_returns = test_data.pct_change().dropna()

# Fit a linear regression model on training data
model = LinearRegression()
model.fit(np.array(train_returns).reshape(-1, 1), train_returns)

# Predict returns for test data
predicted_returns = model.predict(np.array(test_returns).reshape(-1, 1))

# Calculate Mean Absolute Error (MAE)
mae = metrics.mean_absolute_error(test_returns, predicted_returns)

# Calculate Mean Squared Error (MSE)
mse = metrics.mean_squared_error(test_returns, predicted_returns)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```
